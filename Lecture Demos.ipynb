Last but not least, **scipy.optimize** is a module within the SciPy library that can be used to find a best fit line. Why is this useful? Let's say we are doing an experiment to find the coefficient of static fricion of different materials pairing. 

After running the experiment multiple times with differnt sets of weights, we acquire the following data:

```python
x_dat = np.array([0.40, 0.79, 1.19, 1.38, 1.58, 1.97])
y_dat = np.array([0.17, 0.29, 0.40, 0.61, 0.62, 0.81])
y_err = np.array([0.01, 0.02, 0.01, 0.02, 0.04, 0.02])
```

To test whether this data set has any relationship (we expect a linear from $F=μN$), we can **fit** out data to a linear model and **test** the goodness of the fit.

```python
#First, define the model you want to fit to:

def model_lin(x,m):
    return m*x

#Second, calculate the best fit slope and its error. We will denote the slope as m.

lin_opt, lin_cov = opt.curve_fit(model_lin, x_dat, y_dat, sigma=y_err)

m = lin_opt
err = np.diagonal(lin_cov)
```
